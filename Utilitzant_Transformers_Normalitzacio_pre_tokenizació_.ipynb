# **Normalització i pre-tokenització**


---

El tokenizer realitza dos passos: normalització i pre-tokenització.

1. **Normalització** implica una neteja general, com ara eliminar espais en blanc innecessaris, minúscules i/o eliminar accents.
1. **Pre-tokenització** dividir els textos en petites entitats, com ara paraules. Aquí és on entra en joc el pas previ a la tokenització

image.png

```python
from transformers import AutoTokenizer

tokenizer = AutoTokenizer.from_pretrained("bert-base-uncased")
print(type(tokenizer.backend_tokenizer))

#Fast tokenizer ofereixen un accés fàcil a la Normaització del text. Inclosa quan es tokenitza un text.
#Slow tokenizer, pot requerir una costumització del tokenizer (entrenament).
```

```python
print(tokenizer.backend_tokenizer.normalizer.normalize_str("Héllò hôw are ü?"))
```

# **Pre-tokenització**

---
Dividir els textos en petites entitats, com ara paraules. Aquí és on entra en joc el pas previ a la tokenització



```python
tokenizer.backend_tokenizer.pre_tokenizer.pre_tokenize_str("Hello, how are  you?")
```

```python
tokenizer = AutoTokenizer.from_pretrained("gpt2")
tokenizer.backend_tokenizer.pre_tokenizer.pre_tokenize_str("Hello, how are  you?")
```

```python
tokenizer = AutoTokenizer.from_pretrained("t5-small")
tokenizer.backend_tokenizer.pre_tokenizer.pre_tokenize_str("Hello, how are  you?")
```
