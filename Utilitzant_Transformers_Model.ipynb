# **Cear un transformer**


---


El primer que haurem de fer per inicialitzar un model BERT és carregar un objecte de configuració. Això crea un model BERT buit, sense pesos preentrenats.
És com tenir l’estructura de BERT però amb pesos inicialitzats aleatòriament.
Es fa servir per entrenar el model des de zero, o fer experiments amb configuracions personalitzades

```python
from transformers import BertConfig, TFBertModel

# Aplicant la configuració
config = BertConfig()

# Construir el model aplicant la configuració
model = TFBertModel(config)
```
